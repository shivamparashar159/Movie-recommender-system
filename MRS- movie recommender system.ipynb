# We will use content based filtering
This algorithm recommends products that are similar to the ones that a user has liked in the past. This similarity (generally cosine similarity) is computed from the data we have about the items as well as the user’s past preferences.

In genres we only required the name of the genres so we will remove the rest for that we will create a helper function

```python
#first we will convert the string of list into list
import ast
ast.literal_eval('[{"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 28, "name": "Action"}]')
```

```python
# creating a helper function
def convert(obj):
    L= []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
# storing the data into the genres column
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

 In the same manner we will apply this on keywords column

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
movies['cast'][1]
```

we wil take only top 4 cast in the cast column

```python
# creating a function named convert4 
def convert4(obj):
    L= []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter!= 4:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
         
```

```python
# storing the data into the cast column
movies['cast'] = movies['cast'].apply(convert4)
```

```python
movies.head()
```

```python
movies['crew'][1]
```

 we will only take director from the crew column

```python
# creating a function named fetch_director
def fetch_Director(obj):
    L= []
    for i in ast.literal_eval(obj):
        if i['job']== 'Director':
            L.append(i['name'])
            break
    return L
```

```python
# storing the data into the cast column
movies['crew'] = movies['crew'].apply(fetch_Director)
```

```python
movies.head()
```

 we will convert overview column which is a string into list so that we can concatenate it with other list

```python
# using a lambda function
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

we will remove the spaces from genres, keywords, cast column so that it wont recommend us wrong movies as the first name of the persons could be same

```python
# removing space from the columns
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

we will make a new column named tags which will concatenate all the columns apart from title column into one

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

 we will make a new dataframe taking movie_id , title and tags column

```python
new_movies = movies[['movie_id','title','tags']]
```

```python
new_movies.head()
```

we will convert the tags list into string

```python
new_movies['tags'] = new_movies['tags'].apply(lambda x:" ".join(x))
```

```python
new_movies['tags'][0]
```

converting all the words into lower case

```python
new_movies['tags'] = new_movies['tags'].apply(lambda x:x.lower())
```

```python
new_movies.head()
```

we will covert all the text into vectors using text vectorizations-bag of words so that all the movies will be converted into a single vector and then we can recommend similiar movies as the closest vectors to that movie

In bag of words we will concatenate all the words in tags and then we will take out 6000 most frequent words. Now we will take every tag from every movie and find out how many times out of those 6000 words, the words in the tags are occuring. i.e frequency of the words

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=6000, stop_words= 'english')
```

```python
vectors = cv.fit_transform(new_movies['tags']).toarray()
```

```python
vectors
```

```python
cv.get_feature_names()
```

```python
import nltk
```

in the 6000 words there are many similiar words like affection,affections. so we will aplly stemming so that we will only get single word

creating a helper function stemming

```python
from nltk.stem.porter import PorterStemmer
stemmer = PorterStemmer()

def stemming(text):
    stemmed_output = []
    for i in text.split():
        stemmed_output.append(stemmer.stem(i))
    string = " ".join(stemmed_output)
    return string
```

```python
# saving the data in the same column
new_movies['tags'] = new_movies['tags'].apply(stemming)
```

```python
new_movies.head()
```

```python
vectors = cv.fit_transform(new_movies['tags']).toarray()
```

```python
vectors
```

```python
cv.get_feature_names()
```

Now we will find the cosine distance between each vectors of movie. lesser the cosine distance higher the similiarity of the movie

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similiarity = cosine_similarity(vectors)
```

```python
similiarity[1]
```

we will sort the movie according to similiarity. but if we sort it as it is the index position will change 

so for that we will use enumerate function which will hold the position 

```python
sorted(list(enumerate(similiarity[0])),reverse= True, key=lambda x:x[1])[1:6]
```

 we will create a function recommend in which if we provide a movie it will return 5 similiar movie

```python
def recommend(movie):
    movie_index = new_movies[new_movies['title'] == movie].index[0]
    distances = similiarity[movie_index]
    movie_list = sorted(list(enumerate(distances)),reverse= True, key=lambda x:x[1])[1:6]
    
    for i in movie_list:
        print(i[0])
```

```python
recommend('Batman Begins')
```

we will fetch the name of the movie from the index

```python
def recommend(movie):
    movie_index = new_movies[new_movies['title'] == movie].index[0]
    distances = similiarity[movie_index]
    movie_list = sorted(list(enumerate(distances)),reverse= True, key=lambda x:x[1])[1:6]
    
    for i in movie_list:
        print(new_movies.iloc[i[0]].title)
```
